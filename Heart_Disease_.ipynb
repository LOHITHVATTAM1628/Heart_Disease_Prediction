```python
df['trestbps'] = df['trestbps'].replace(0, df['trestbps'].mean())
df['chol'] = df['chol'].replace(0, df['chol'].mean())
df['thalach'] = df['thalach'].replace(0, df['thalach'].mean())
df['oldpeak'] = df['oldpeak'].replace(0, df['oldpeak'].median())
df['ca'] = df['ca'].replace(0, df['ca'].median())
df['thal'] = df['thal'].replace(0, df['thal'].median())
```

```python
df.tail(100)
```

```python
x=df.drop(columns='target',axis=1)
y=df['target']
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.boxplot(data=x,ax=ax)
plt.show()
```

```python
cols=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach','exang', 'oldpeak', 'slope', 'ca', 'thal',]
for col in cols:
  q1 = x[col].quantile(0.25)
  q3 = x[col].quantile(0.75)
  IQR = q3 - q1
  lower_bound = q1 - 1.5*IQR
  upper_bound = q3 + 1.5*IQR
  mask = (x[col] >= lower_bound) & (x[col] <= upper_bound)
```

```python
x_out=x[mask]
y_out=y[mask]
```

```python
x_out.shape
```

```python
y_out.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x_out)
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x_scaled,ax=ax)
plt.savefig('boxplot.jpg')
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=cols)
x_scaled.describe()
```

```python
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=x_scaled,ax=ax)
plt.savefig('boxplot.jpg')
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_out,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
y_train.value_counts()
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
cla=LogisticRegression()
cla.fit(x_train,y_train)
```

```python
y_pred=cla.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
t=['Heart disease','No Heart disease']
print(classification_report(y_test,y_pred,target_names=t))
```
